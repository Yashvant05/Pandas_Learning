```
df['episodes'] = df['Title'].str.extract(r'(\(\d+\s*eps\))')
```

```
print(df['episodes'].to_string(index=False))
```

```
def extract_eps(txt):
  check = False
  data = ""
  for i in txt:
    if i == ')':
      check = False
    if check == True:
      data = data + i
    if i == '(':
      check = True
  return data  # ✅ now it works
```

```
df["Episodes"] = df['Title'].apply(extract_eps)
```

```
df["Episodes"] = df["Episodes"].str.replace("eps", "")
```

```
def extract_time(txt):
   check = False
   data = ""
   for i in range(len(txt)):
     if txt[i] == ")":
        for j in range(i+1, i+20):
         data += txt[j]
        return data
```

```
df["Total Time"] = df["Title"].apply(extract_time)
```

```
df
```

```
df
```

```
df['start'] = pd.to_datetime(df['Total Time'].str.split(' - ').str[0])
df['end'] = pd.to_datetime(df['Total Time'].str.split(' - ').str[1])

df['months'] = (df['end'].dt.year - df['start'].dt.year) * 12 + (df['end'].dt.month - df['start'].dt.month)
```

```
df["Months"] = df['months'] + 1
```
