```
# Group By Aggregation
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B'],
    'Store': ['S1', 'S1', 'S2', 'S2', 'S1', 'S2', 'S2', 'S1'],
    'Sales': [100, 200, 150, 250, 120, 180, 200, 300],
    'Quantity': [10, 15, 12, 18, 8, 20, 15, 25],
    'Date': pd.date_range('2023-01-01', periods=8)
}
df = pd.DataFrame(data)
```

```
df
```

```
 # Group by category and calc the sum of sales
 cat = df.groupby('Category')['Sales'].sum()
 cat
```

```
# Group by store and calculate the sum of sales
store = df.groupby('Store')['Sales'].sum()
store
```

```
# Group by multiple columns
# Group by Category and Stores
g = df.groupby(['Category', 'Store'])['Sales'].sum()
g
```

```
# Aggregation - Combine multiple things and get one answer
df['Sales'].mean()
```

```
h = df['Sales'].agg(['sum', 'count', 'mean', 'median', 'min', 'max', 'std'])
h
```

***Pivot Tables***

```
data = {
    'Date': pd.date_range('2024-01-01', periods=20),

    'Product': [
        'Laptop', 'Mobile', 'Tablet', 'Headphones'
    ] * 5,

    'Region': [
        'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West'
    ],

    'Sales': np.random.randint(100, 1000, 20),

    'Units': np.random.randint(1, 10, 20),

    'Salesperson': [
        'Rahul', 'Priya', 'Amit', 'Neha'
    ] * 5
}

df = pd.DataFrame(data)

df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
```

```
df
```

```
pd.pivot_table(df, values = 'Sales', index = 'Region', columns = 'Product', aggfunc = 'max')
```

```
pivot2 = pd.pivot_table(df, values = ['Sales', 'Units'], index = 'Region', columns = 'Product', aggfunc = 'sum')
```

```
pivot2
```

**Cross Table**

```
pd.crosstab(df['Region'], df['Product'])
```
